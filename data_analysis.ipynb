```python
import psycopg2 as pg2
import matplotlib.pyplot as plt
from config import password_sql
```

```python
conn = pg2.connect(database='employee_sql_homework', user='postgres', password=password_sql)
```

```python
cur = conn.cursor()
```

```python
cur.execute('SELECT salary FROM salaries')
```

```python
data=cur.fetchall()
```

```python
salary_list=[]
for (x,) in data:
    salary_list.append(x)
```

```python
plt.hist(salary_list)
plt.xlabel('Salary')
plt.ylabel('Counts')
plt.show()
```

```python
cur.execute('SELECT titles.title, AVG(salaries.salary) FROM employees JOIN salaries ON employees.emp_no=salaries.emp_no JOIN titles ON employees.emp_title_id=titles.title_id GROUP BY titles.title')
```

```python
data=cur.fetchall()
```

```python
titles=[]
salary_average=[]

for (x,y) in data:
    titles.append(x)
    salary_average.append(y)
```

```python
plt.bar(titles, salary_average, color='r', alpha = 0.5, align = "center")
plt.xticks(rotation='vertical')
plt.show()
```
